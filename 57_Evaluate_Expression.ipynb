# Evaluate Expression

Problem Description
 
 

An arithmetic expression is given by a charater array A of size N. Evaluate the value of an arithmetic expression in Reverse Polish Notation.

Valid operators are +, -, , /. Each character may be an integer or an operator.



Problem Constraints
1 <= N <= 105



Input Format
The only argument given is character array A.



Output Format
Return the value of arithmetic expression formed using reverse Polish Notation.



Example Input
Input 1:
    A =   ["2", "1", "+", "3", ""]
Input 2:
    A = ["4", "13", "5", "/", "+"]


Example Output
Output 1:
    9
Output 2:
    6


Example Explanation
Explaination 1:

    starting from backside:
    
    * : () * ()
    
    3 : () * (3)
    
    + : (() + ()) * (3)
    
    1 : (() + (1)) * (3)
    
    2 : ((2) + (1)) * (3)
    
    ((2) + (1)) * (3) = 9
    
Explaination 2:

    + : () + ()
    
    / : () + (() / ())
    
    5 : () + (() / (5))
    
    13 : () + ((13) / (5))
    
    4 : (4) + ((13) / (5))
    
    (4) + ((13) / (5)) = 6

```python
class Solution:
    # @param A : list of strings
    # @return an integer
    def evalRPN(self, A):
        operations = []
        numbers = []
        num_ops = 0
        while len(A) > 0:
            print("A:", A)
            print('numbers:',numbers)
            print('ops:', operations)
            item = A.pop()
            if item in ['+', '-', '*', '/']:
                # its an operation!!
                operations.append(item)
                # reset number of ops
                num_ops = 0
            else:
                # its a number!!
                numbers.append(int(item))
                # increase number of ops
                num_ops += 1
            if num_ops == 2:
                # its time to calculate the result! :D
                op = operations.pop()
                num1 = numbers.pop()
                num2 = numbers.pop()
                if op == '+':
                    result = num1 + num2
                elif op == '-':
                    result = num2 - num1
                elif op == '*':
                    result = num2 * num1
                else:
                    result = num2//num1
                if len(A) != 0:
                    A.append(str(result))
                else:
                    numbers.append(int(result))
        while len(operations) > 0:
            op = operations.pop()
            num1 = numbers.pop()
            num2 = numbers.pop()
            if op == '+':
                result = num1 + num2
            elif op == '-':
                result = num2 - num1
            elif op == '*':
                result = num2 * num1
            else:
                result = num2//num1
            numbers.append(int(result))
        print("A:", A)
        print('numbers:',numbers)
        print('ops:', operations)
        return(numbers[-1])

```

```python
class Solution:
    # @param A : list of strings
    # @return an integer
    def evalRPN(self, A):
        stack = []
        ret = 0
        first = 0
        second = 0
        for a in A:
            if a == '+' or a == '-' or a == '/' or a == '*':
                if len(stack) < 2:
                    return 0
                second = stack.pop()
                first = stack.pop()
                if a == '+':
                    ret = first + second
                    stack.append(ret)
                elif a == '-':
                    ret = first - second
                    stack.append(ret)
                elif a == '*':
                    ret = first * second
                    stack.append(ret)
                elif a == '/':
                    ret = first / second
                    stack.append(ret)
            else:
                stack.append(int(a))
        if len(stack) == 1:
            return stack[0]
        else:
            return 0
            
```
