```python
class Solution:
    
    occurs = []
    
    def checkOccurrence(self, A, s):
        i = 0
        while i < len(A) - len(s):
            j = 0
            k = i
            while k < len(A) and j < len(s) and A[k] == s[j]:
                if j == len(s) - 1:
                    self.occurs.append(i)
                j+=1
                k+=1
            i += 1
    
    def generateConcatenations(self, A, B):
        subs = []
        if len(B) == 0:
            return []
        if len(B) == 1:
            return [B[0]]
        for i in range(len(B)):
            s = B[i]
            newB = B.copy()
            del newB[i]
            concs = self.generateConcatenations(A,newB)
            for c in concs:
                if len(s+c) < len(A):
                    subs.append(s + c)
                    self.checkOccurrence(A, s+c)
                    print(s+c)
        return subs
    # @param A : string
    # @param B : tuple of strings
    # @return a list of integers
    def findSubstring(self, A, B):
        self.occurs = []
        self.pos = []
        self.generateConcatenations(A,list(B))
        self.occurs = list(set(self.occurs))
        self.occurs.sort()
        return self.occurs
        
        
```
