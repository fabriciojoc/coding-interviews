# Level Order

## Problem Description

Given a binary tree, return the level order traversal of its nodes' values. (ie, from left to right, level by level).

## Input Format

First and only argument is root node of the binary tree, A.

## Output Format

Return a 2D integer array denoting the zigzag level order traversal of the given binary tree.

## Example Input

Input 1:

```
    3      
   / \     
  9  20    
    /  \   
   15   7  
```

Input 2:

```
   1    
  / \   
 6   2  
    /   
   3    
```

## Example Output

Output 1:

```
 [          
   [3],     
   [9, 20], 
   [15, 7]  
 ]          
```

Output 2:

```
 [         
   [1]     
   [6, 2]  
   [3]     
 ]         
```

## Example Explanation

Explanation 1:

Return the 2D array. Each row denotes the traversal of each level.



````python
# Definition for a  binary tree node
# class TreeNode:
#	def __init__(self, x):
#		self.val = x
#		self.left = None
#		self.right = None

class Solution:

    def levelOrderAux(self, A,level):
        if not A:
            return
        if level not in self.levels.keys():
            self.levels[level] = [A.val]
        else:
            self.levels[level].append(A.val)
        self.levelOrderAux(A.left, level+1)
        self.levelOrderAux(A.right, level+1)

    # @param A : root node of tree
    # @return a list of list of integers
    def levelOrder(self, A):
        self.levels = {}
        self.levelOrderAux(A,0)
        result = []
        for k,v in self.levels.items():
            result.append(v)
        return result

````
